```python
# 绘制促销商品和非促销商品销售金额的周环比增长率柱状图
data.head()
```

```python
Promotion = data.loc[data['是否促销']=='是',:].groupby('周数')['销售金额'].sum()
```

```python
No_Promotion = data.loc[data['是否促销']=='否',:].groupby('周数')['销售金额'].sum()
```

```python
Promotion.head()
```

```python
Prate = []
for i in range(2,19):
    Prate.append((Promotion[i]-Promotion[i-1])/Promotion[i])
```

```python
len(Prate)
```

```python
NPrate = []
for i in range(2,19):
    NPrate.append((No_Promotion[i]-No_Promotion[i-1])/No_Promotion[i])
```

```python
x = np.arange(len(Prate))
width = 0.35
fig,ax = plt.subplots(figsize=(7,10))
ax.bar(x-width/2,Prate,width,label = '促销')
ax.bar(x+width/2,NPrate,width,label = '非促销')
ax.set_ylabel('比率')
ax.set_title('促销商品和非促销商品销售金额的周环比增长率')
ax.set_xticks(x)
ax.grid()
ax.legend()
plt.show()
```
