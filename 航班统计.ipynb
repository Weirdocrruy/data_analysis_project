```python
data['cabinRemain'].value_counts()
```

```python
def swapint(x):
    if x=='9张+':
        return 10
    elif x=='剩9张':
        return 9
    elif x=='剩8张':
        return 8
    elif x=='剩7张':
        return 7
    elif x=='剩6张':
        return 6
    elif x=='剩5张':
        return 5
    elif x=='剩4张':
        return 4
    elif x=='剩3张':
        return 3
    elif x=='剩2张':
        return 2
    elif x=='剩1张':
        return 1
```

```python
data['cabinRemain'] = data['cabinRemain'].apply(swapint)
```

```python
price['point'] = price.index.str.split('--')

price['Startpoint'] = price['point'].str[0]

price['endpoint'] = price['point'].str[1]

price['deptCityName'] = price['Startpoint'].str.split('-').str[1]

price['destCityName'] = price['endpoint'].str.split('-').str[1]

price.index = np.arange(price.shape[0])

price['number'] = np.arange(price.shape[0])+1
```
