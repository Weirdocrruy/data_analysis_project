```python
import os
import pandas as pd
import numpy as np
import re
import sklearn
```

```python
def readtxt(file_path):
    file_path = file_path
    file_names = os.listdir(file_path)
    file_name = [file_path + '/' + i for i in file_names]
    main_name = re.findall('.{%d}/SH#0[0-9]{5}.txt'%(len(file_path)),' '.join(file_name),re.S)
    file_name.remove(main_name[0])
    data = {}
    for i in file_name:
        data_ = pd.read_csv(i,sep='\t',encoding='gb18030')
        data.update({data_.reset_index().columns[6].split()[1]:data_.reset_index()['level_4'][1:-1]})
    data = pd.DataFrame(data)
    data = pd.concat([pd.read_csv(main_name[0],sep='\t',encoding='gb18030').reset_index()[['level_0','level_4']][1:-1],data],axis=1)
    data.rename(columns = {'level_0':'日期','level_4':'上证50'},inplace=True)
    data.to_csv('上证50.csv')
    return data
```

```python
def get_data_shang(file_path):
    file_names = os.listdir(file_path)
    file_name = [file_path + '/' + i for i in file_names]
    file_name.pop(1)
    file_name
    data = []
    for i in file_name:
        data.append(pd.read_csv(i,sep='\t',encoding='gb18030',skiprows=[0,1,142],usecols=[*range(1,7)],header=None))
    data = pd.concat([pd.DataFrame(data[0]),pd.DataFrame(data[1]),pd.DataFrame(data[2]),pd.DataFrame(data[3]),pd.DataFrame(data[4])],axis=0)
    return data
```
